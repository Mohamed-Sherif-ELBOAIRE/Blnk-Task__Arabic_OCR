```python
# Define New Output Folders
Blurred_Grayscaled =   'D://Mohamed Sheriff//Projects//Blnk Task//Dataset//Blurred_Grayscaled//'
Sharpened_Grayscaled = 'D://Mohamed Sheriff//Projects//Blnk Task//Dataset//Sharpened_Grayscaled//'

# Create output directories if they don't exist
os.makedirs(Blurred_Grayscaled, exist_ok=True)
os.makedirs(Sharpened_Grayscaled, exist_ok=True)


# Function to convert an RGB image to grayscale
def rgb_to_gray(rgb_image):
    r, g, b = rgb_image[:, :, 0], rgb_image[:, :, 1], rgb_image[:, :, 2]
    gray = (0.2989 * r) + (0.5870 * g) + (0.1140 * b)
    return gray


# Intialize and Loop through each image in the input directories
blur_files    = [f for f in blur_item if f.endswith('.jpg')]
sharpen_files = [f for f in sharp_item if f.endswith('.jpg')]

# ##############################################
# # Loop for Blurred Images
# ##############################################
for image_file in blur_files:
    # Construct the full path to the input image
    input_path = os.path.join(blurred, image_file)

    # Load the image using cv2
    input_img = cv2.imread(input_path)

    # Check if the image was loaded successfully
    if input_img is not None:
        # Convert the RGB image to grayscale
        gray_image = cv2.cvtColor(input_img, cv2.COLOR_BGR2GRAY)
        resized_image = cv2.resize(gray_image, (370, 90))

        # Save the grayscale image to the output directory
        output_path = os.path.join(Blurred_Grayscaled, f'gray_{image_file}')
        cv2.imwrite(output_path, resized_image)
    else:
        print(f"Failed to load image: {input_path}")   
##############################################
# Loop for Sharpened Images
##############################################
for image_file in sharpen_files:
    # Construct the full path to the input image
    input_path = os.path.join(sharpened, image_file)

    # Load the image using cv2
    input_img = cv2.imread(input_path)

    # Check if the image was loaded successfully
    if input_img is not None:
        # Convert the RGB image to grayscale
        gray_image = cv2.cvtColor(input_img, cv2.COLOR_BGR2GRAY)
        resized_image = cv2.resize(gray_image, (370, 90))

        # Save the grayscale image to the output directory
        output_path = os.path.join(Sharpened_Grayscaled, f'gray_{image_file}')
        cv2.imwrite(output_path, resized_image)
    else:
        print(f"Failed to load image: {input_path}") 
```

```python
# Intialize and Apply Gaussian blur for noise reduction with Binary/NiBlack Thresholds
kernel_size = (5,5)  # Adjust the kernel size as needed
sigma_x = 0           # Adjust the standard deviation as needed


# Define New Output Folders -  Binary Threshold
Binary_sharp = 'D://Mohamed Sheriff//Projects//Blnk Task//Dataset//Binary_Sharp//'

# Create output directories if they don't exist
os.makedirs(Binary_sharp, exist_ok=True)



# Intialize and Loop through gray scaled images to apply Binary threshold Threshold upon Sharped dataset
Binary_sharp_dataset = [f for f in sharp_gray if f.endswith('.jpg')]

for filename in Binary_sharp_dataset:

    # Load the grayscale image
    image = imread(os.path.join(Sharpened_Grayscaled, filename))
    
    # Apply Gaussian blur
    Gaussian = cv2.GaussianBlur(image, kernel_size, sigma_x)
    # Apply binary thresholding
    threshold_value = 128  # Set your threshold value
    _, binary_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)

    # Save the binary thresholded image
    binary_output_path = os.path.join(Binary_sharp, f'binary_sharp_{filename}')
    cv2.imwrite(binary_output_path, binary_image)
```

```python
# Define New Output Folders -  NiBlack Threshold
NiBlack_sharp = 'D://Mohamed Sheriff//Projects//Blnk Task//Dataset//NiBlack_Sharp'

# Create output directories if they don't exist
os.makedirs(NiBlack_sharp, exist_ok=True)

# Intialize and Loop through gray scaled images to apply Niblack threshold Threshold upon Sharped dataset
NiBlack_sharp_dataset = [f for f in sharp_gray if f.endswith('.jpg')]

for filename in NiBlack_sharp_dataset:

    # Load the grayscale image
    image = cv2.imread(os.path.join(Sharpened_Grayscaled, filename), cv2.IMREAD_GRAYSCALE)
    
    # Apply Gaussian blur
    Gaussian = cv2.GaussianBlur(image, kernel_size, sigma_x)
    
    # Define the size of the local neighborhood (window size)
    window_size = 25

    # Define the Niblack thresholding constant 'k'
    k = -0.4

    # Apply Niblack thresholding
    Ni_image = cv2.ximgproc.niBlackThreshold(image, maxValue=255, type=cv2.THRESH_BINARY, blockSize=window_size, k=k)
    
    # Save the Niblack thresholded image
    niblack_output_path = os.path.join(NiBlack_sharp, f'niblack_sharp_{filename}')
    cv2.imwrite(niblack_output_path, Ni_image)
```

```python
# Outlining the model/architecture of our CNN
model = Sequential()
   
model.add(Conv2D(64, kernel_size=(3, 3), padding='same', input_shape=(90, 370, 1), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(128, kernel_size=(5, 5), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))

model.add(Conv2D(256, kernel_size=(3, 3), padding='same', activation='relu'))
model.add(Conv2D(256, kernel_size=(3, 3), padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
    
model.add(Conv2D(512, kernel_size=(3, 3), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
    
model.add(Conv2D(512, kernel_size=(3, 3), padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(512, kernel_size=(3, 3), padding='same', activation='relu'))

model.add(Flatten())
    
# Bidirectional LSTM layers for sequence recognition
Bidirectional(LSTM(128, return_sequences=True, dropout=0.2))
Bidirectional(LSTM(128, return_sequences=True, dropout=0.2))
    
# Output layer with CTC loss

output = Dense(len(char_set) + 1, activation='softmax')

model.add(output)
```

```python
def CTCLoss(y_true, y_pred):
    # Compute the training-time loss value
    batch_len = tf.cast(tf.shape(y_true)[0], dtype="int64")
    input_length = tf.cast(tf.shape(y_pred)[1], dtype="int64")
    label_length = tf.cast(tf.shape(y_true)[1], dtype="int64")

    input_length = input_length * tf.ones(shape=(batch_len, 1), dtype="int64")
    label_length = label_length * tf.ones(shape=(batch_len, 1), dtype="int64")
    print(input_length,label_length)
    loss = keras.backend.ctc_batch_cost(y_true, y_pred, input_length, label_length)
    return loss
```

```python
# Compile the model
model.compile(optimizer = 'Adam' , loss = CTCLoss, metrics=["accuracy"])
```

```python
log_dir = "D://Mohamed Sheriff//Projects//Blnk Task//Logs//" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir, histogram_freq=1)
```

```python
# Defining early stopping to prevent overfitting
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor = 'val_loss',
    mode = 'auto',
    min_delta = 0,
    patience = 2,
    verbose = 0, 
    restore_best_weights = True
    )
```

```python
# Train the model
model.fit(x_train_gray, y_train, 
        epochs=10,
        validation_data=(x_test_gray, y_test),
        callbacks=[tensorboard_callback, early_stopping])
```
